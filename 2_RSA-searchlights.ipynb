# Make some masks

```python
hpc_img_fname = f'{mask_dir}/Repl_Bilateral_HPC_harvardoxford_maxprob-thr25-1mm.nii'
hpc_img = nib.load(hpc_img_fname)
hpc_img_data = hpc_img.get_fdata()

# resample to hpc mask
gm_img = nilearn.datasets.load_mni152_gm_mask()
gm_img = nilearn.image.resample_to_img(gm_img, hpc_img, interpolation='nearest') 
gm_img.to_filename(f'{mask_dir}/Repl_GM.nii')
gm_img_data = gm_img.get_fdata()

# negate hpc mask
gm_img_data_ = np.where(hpc_img_data==1, 0, gm_img_data)
gm_img_nohpc = nilearn.image.new_img_like(gm_img, gm_img_data_)
nilearn.plotting.plot_roi(gm_img_nohpc)
gm_img_nohpc.to_filename(f'{mask_dir}/Repl_GM_without_Bilateral_HPC.nii')
```

# Non parametric tests

```python
sl_dir   = f'{base_dir}/lsa_onset_128hpf/searchlights'
which    = 'dimension-ps_cube5mm'
sl_dict  = find_sl_imgs(which,  sl_dir=f'{sl_dir}/nifties')

mask_dict = {'L_HPC': f'{mask_dir}/Repl_L_HPC_harvardoxford_maxprob-thr25-1mm.nii',
             'Bilateral_HPC': f'{mask_dir}/Repl_Bilateral_HPC_harvardoxford_maxprob-thr25-1mm.nii',
             'GM': f'{mask_dir}/Repl_GM.nii',
             'GM_without_L_HPC': f'{mask_dir}/Repl_GM_without_L_HPC.nii',
             'GM_without_Bilateral_HPC': f'{mask_dir}/Repl_GM_without_Bilateral_HPC.nii'}

overwrite = False
for mask in mask_dict.keys():
    for fwhm in [8]:
        perm_ttest_imgs, thr_imgs = {}, {}

        for sample in ['Combined']:
            sample_n = len(sl_dict[sample]['sub_ids']) 
            fname = f'{sl_dir}/{which}_perm_ttest_{mask}_{fwhm}fwhm_{sample}_n{sample_n}.nii.gz'
            if os.path.exists(fname) and not overwrite:
                print(f'Loading {fname}')
                perm_ttest_imgs[sample] = nib.load(fname)
            else:
                perm_ttest_imgs[sample] = compute_permutation_ttest(sl_dict[sample]['imgs'], 
                                                                    mask_img=mask_dict[mask], 
                                                                    n_perm=1000, 
                                                                    fwhm=fwhm, 
                                                                    tfce=False)
                perm_ttest_imgs[sample].to_filename(fname)

            thr_imgs[sample], table = threshold_neglog_img(perm_ttest_imgs[sample], 
                                                           alpha=0.05, 
                                                           cluster_threshold=0, 
                                                           vox_size=2.1)
            nilearn.plotting.plot_stat_map(thr_imgs[sample], display_mode='ortho', draw_cross=False, vmax=5)                    
            plt.show()
            print_df(table)
```

### Samples combined w/ controls

```python
# collect info
sample_imgs, sub_dfs = [], []
for s, sample in enumerate(['Initial', 'Validation']):
    sub_ids = [s.split('/')[-1].split('_')[0] for s in sl_dict[sample]['imgs']]
    sub_dfs.append(pd.DataFrame({'subject_label': sub_ids, 'sample': s}))
    sample_imgs.extend(sl_dict[sample]['imgs'])
sub_df = pd.concat(sub_dfs)

# run glm for both samples, contrllling for sample
design_matrix = nilearn.glm.second_level.make_second_level_design_matrix(sub_df['subject_label'].values, confounds=sub_df)
glm = SecondLevelModel(mask_img=None, smoothing_fwhm=6, n_jobs=-1)
glm = glm.fit(sample_imgs, design_matrix=design_matrix) 
print_df(glm.design_matrix_.head(3))

stat_img = glm.compute_contrast(second_level_contrast='intercept', output_type='z_score')
nilearn.plotting.view_img(stat_img)

# gm_img = nilearn.masking.compute_multi_brain_mask(sample_imgs, threshold=0.25, connected=False, opening=2, mask_type='gm')
thr_stat_img = threshold_second_level_img(stat_img, mask_img=None, two_sided=False, 
                                          alpha=0.05, cluster_extent=0, height_control='fdr')
nilearn.plotting.view_img(thr_stat_img)
```

### Parametric t-tests

```python
ttest_imgs = {
    sample: compute_ttest(sl_dict[sample]['imgs'], fwhm=fwhm)
    for sample in ['Combined']
}
thr_imgs = {'Initial': {}, 'Validation': {}, 'Combined': {}}
for sample in ['Combined']:
    thr_imgs[sample]['HPC'] = threshold_second_level_img(ttest_imgs[sample], mask_img=hpc_img, 
                                                         two_sided=False, alpha=0.05, 
                                                         cluster_extent=0, height_control='fpr')
    thr_imgs[sample]['wholebrain'] = threshold_second_level_img(ttest_imgs[sample], mask_img=None,
                                                                two_sided=False, alpha=0.001,
                                                                cluster_extent=50, height_control='bonferroni')
    
for sample in ['Combined']:
nilearn.plotting.plot_stat_map(thr_imgs[sample]['HPC'], draw_cross=False, cut_coords=[-32, -14, -16])
plt.show()
nilearn.plotting.plot_stat_map(thr_imgs[sample]['wholebrain'], draw_cross=False)
plt.show()
```
